```python
import os
import random
import skimage.data
import skimage.transform
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


# Allow image embeding in notebook
%matplotlib inline
```

```python
def load_data(data_dir):

    # findet alle Unterordner vom data_dir. Jedes einzelne ist im endeffekt eine Klasse
    directories = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]
    labels = []
    images = []
    for d in directories:
        label_dir = os.path.join(data_dir, d)
        file_names = [os.path.join(label_dir, f) 
                      for f in os.listdir(label_dir) if f.endswith(".ppm")]
    
        for f in file_names:
            images.append(skimage.data.imread(f))
            labels.append(int(d))
    return images, labels


train_data_dir = os.path.join("./GTSRB/GTSRB/Final_Training/Images")
test_data_dir = os.path.join("./GTSRB/GTSRB/Final_Test/Images")

images, labels = load_data(train_data_dir)
```

```python
print "Verschiedene Klassen: " , len(set(labels))
print "Anzahl Gesamt Bilder: ", len(images)
```

```python
def display_images_and_labels(images, labels):
    """Ploten vom ersten Bild in den einzelnen Klassen/labels"""
    unique_labels = set(labels)
    plt.figure(figsize=(15, 15))
    i = 1
    for label in unique_labels:

        image = images[labels.index(label)]
        plt.subplot(8, 8, i)  #8 rows , 8 columns
        plt.axis('off')
        plt.title("Label {0} ({1})".format(label, labels.count(label)))
        i += 1
        _ = plt.imshow(image)
    plt.show()

display_images_and_labels(images, labels)
```

```python
def display_label_images(images, label):
    """Plotten von Bildern des angegeben Labels"""
    limit = 50  # 50 images
    plt.figure(figsize=(15, 5))
    i = 1

    start = labels.index(label)
    end = start + labels.count(label)
    for image in images[start:end][:limit]:
        plt.subplot(5, 10, i)  # 5 rows, 10 per row
        plt.axis('off')
        i += 1
        plt.imshow(image)
    plt.show()

display_label_images(images, 26)
```
